```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv('../data/creditcard.csv')

print(df.shape)

df.head()
```

```python
#Check for values in the 'Class' column
print(df['Class'].value_counts()) 

plt.figure(figsize=(20,15))
plt.suptitle('Exploratory Data Analysis', fontsize=16)

#Visualize the distribution of the 'Amount' feature
plt.subplot(3,2,1)
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')

#Visualize the relationship between 'Amount' and 'Class'
plt.subplot(3,2,2)
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amount by Class')

#Visualize the distribution of the 'Time' feature
plt.subplot(3,2,3)
sns.histplot(df['Time'], bins=50, kde=True)
plt.title('Distribution of Transaction Time')

#Visualize the relationship between 'Time' and 'Class'
plt.subplot(3,2,4)
sns.boxplot(x='Class', y='Time', data=df)
plt.title('Transaction Time by Class')

#Visualize the correlation between features
plt.subplot(3,2,5)
sns.heatmap(df.corr(), annot=False)
```

```python
#Check for missing values
print(df.isnull().sum())

#Check for duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
#Check scale of features
print(df.describe())
```
